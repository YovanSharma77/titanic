```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
test_df.isnull().sum()
```

```python
train_df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
test_df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
print(train_df.isnull().sum())
print('-'*40)
print(test_df.isnull().sum())
```

```python
from sklearn.impute import KNNImputer
features = ['Age','Fare','SibSp','Parch']
train_df_impute, test_df_impute = train_df[features], test_df[features]

imputer = KNNImputer(n_neighbors=5)
train_df_imputed, test_df_imputed = imputer.fit_transform(train_df_impute), imputer.fit_transform(test_df_impute)

train_df['Age'] = train_df_imputed[:, 0]
test_df['Age'] = test_df_imputed[:, 0]
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0],inplace=True)
test_df['Fare'].fillna(test_df['Fare'].mean(),inplace=True)
```

```python
train_df
```

```python
print(train_df.isnull().sum())
print('-'*40)
print(test_df.isnull().sum())
```

```python
train_df.to_csv('train_cleaned.csv',index=False)
test_df.to_csv('test_cleaned.csv',index=False)
```
